#1. Import Library

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#2. Persiapkan Gambar

```
#path gambar
img_path = "Gambar.jpeg"

#baca citra dalam format grayscale
img_gray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

print("Ukuran Citra : ", img_gray.shape)

# Tampilkan citra grayscale
plt.figure(figsize=(5,5))
plt.imshow(img_gray, cmap='gray')
plt.title("Citra Grayscale Asli")
plt.axis('off')
plt.show()
```

#3. Penerapan Thereshold Global Sederhana

```
#Tentukan nilai Threshold
T = 1 #nilai tengah dicoba terlebih dahulu

#Terapkan Thereshold biner sederhana
ret, img_thresh = cv2.threshold(img_gray, T, 255, cv2.THRESH_BINARY)

print("Nilai Threshold yg digunakan = ", ret)

# Tampilkan citra asli dan hasil thresholding berdampingan
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Citra Grayscale Asli")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_thresh, cmap='gray')
plt.title(f"Thresholding Biner (T = {T})")
plt.axis('off')

plt.tight_layout()
plt.show()
```

#4. Mencoba beberapa nilai T

```
threshold_values = [50, 100, 150, 200]
images = []

for T in threshold_values:
    _, th = cv2.threshold(img_gray, T, 255, cv2.THRESH_BINARY)
    images.append((T, th))

plt.figure(figsize=(12,6))
for i, (T, th) in enumerate(images):
    plt.subplot(2, 2, i+1)
    plt.imshow(th, cmap='gray')
    plt.title(f"T = {T}")
    plt.axis('off')

plt.suptitle("Perbandingan Hasil Thresholding untuk Berbagai Nilai T", fontsize=14)
plt.tight_layout()
plt.show()
```
